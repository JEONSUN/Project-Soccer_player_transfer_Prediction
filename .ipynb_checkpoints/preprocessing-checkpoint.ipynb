## 개요

선수의 시장가격과 능력치 사이의 상관성이 존재하는가?

이를 알아보기 위해 FIFA에서 제공하는 데이터를 바탕으로 선수들의 가치를 확인해보고자 한다.

**평가 방식 - RMSE 점수**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
test_data = pd.read_csv('./data/FIFA_test.csv')
train_data = pd.read_csv('./data/FIFA_train.csv')
```

### FIFA_train.csv


|변수명           | 변수 설명                                          |
|------------------|:--------------------------------------------------:|
|**id**                | 선수 고유의 아이디                                 |
|**name**             | 이름                                               |
|**age**              | 나이                                               |
|**continent**        | 선수들의 국적이 포함되어 있는 대륙                 |
|**contract_until**   | 선수의 계약기간                                    |
|**position**         | 선수가 선호하는 포지션 ex) 공격수, 수비수 등       |
|**prefer_foot**      | 선수가 선호하는 발 ex) 오른발                      |
|**reputation**       | 선수가 유명한 정도 ex) 높은 수치일 수록 유명한 선수|
|**stat_overall**     | 선수의 현재 능력치                                 |
|**stat_potential**   | 선수가 경험 및 노력을 통해 발전할 수 있는 정도     |
|**stat_skill_moves** | 선수의 개인기 능력치                               |
|**value**            | FIFA가 선정한 선수의 이적 시장 가격 (단위 : 유로)  |

### Corr

```python
train_corr = train_data.corr()

train_corr.round(2)
```

```python
import seaborn as sns

f, ax = plt.subplots(figsize=(8,6))
plt.title("Data Correlation", fontsize=30, position=(0.5, 1.0+0.05))
sns.heatmap(train_corr, annot=True, fmt='.2f', linewidths=.5, ax=ax, cmap='YlGnBu')
plt.show()
```

```python
# 

plt.plot('reputation','stat_overall',data = train_data)
plt.title('Reputation and Stat_overall Relation')
plt.xlabel('Reputation')
plt.ylabel('stat_ovarall')

plt.show()
```

```python
plt.figure(figsize = (8,6))

plt.scatter('age','value',data = train_data)

plt.show()
```

```python
from bokeh.layouts import column
from bokeh.plotting import figure, show, output_file
from bokeh.transform import jitter

age = train_data['age']

p1 = figure(plot_width=2000, plot_height=1000, title="Age and value")
p1.xgrid.grid_line_color = None
p1.xaxis.ticker = age
p1.circle(x='age', y='value', size=15, alpha=0.4, source=train_data)

p2 = figure(plot_width=2000, plot_height=1000, title="Age and Value")
p2.xgrid.grid_line_color = None
p2.xaxis.ticker = age
p2.circle(x=jitter('age', 0.4), y='value', size=15, alpha=0.4, source=train_data)

output_file("jitter.html")

show(column(p1, p2))
```

### pivot table

```python
# 이름별 고유값 찾기

train_data['name'].unique()
```

```python
# 나이 간략하게 범주화 시킬것

pos_pivot = train_data.pivot_table(values = 'value' , index = 'position', columns = 'age', aggfunc = np.mean, fill_value = 0)

pos_pivot.round(2)
```

```python
# 스텟 간략하게 범주화 시킬것

sta_pivot = train_data.pivot_table(values = 'value' , index = 'position', columns = 'stat_overall', aggfunc = np.mean, fill_value = 0)

sta_pivot
```

```python
contract_df = train_data.groupby('contract_until').sum()
contract_df
```

```python
# 남은 계약기간 총합
con_value = train_data['contract_until'].value_counts()
con_value
```

```python
# 이거 간략하게 할 것

con_value.plot(kind = 'bar')
plt.show()
```
