## random forest를 사용한 예측

#### 대륙과 포지션, 주로 사용하는 발 범주형 변수를 연속형 변수로 변환

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
train_df = pd.read_csv('./data/FIFA_train.csv')
test_df =  pd.read_csv('./data/FIFA_test.csv')
```

```python
train_df.head(10)
```

```python
# 범주형 변수들의 고유값 출력
print(train_df['continent'].unique())
print("-------------------------------------")
print(train_df['position'].unique())
print("-------------------------------------")
print(train_df['prefer_foot'].unique())
```

```python
train_df['continent'] = train_df['continent'].map({'south america' : 0, 'europe' : 1,'africa'  : 2,'asia':3, 'oceania':4})
train_df['position'] = train_df['position'].map({'ST':0, 'GK':1 ,'DF':2 ,'MF':3})
train_df['prefer_foot'] = train_df['prefer_foot'].map({'left':0 ,'right':1})
```

```python
train_df.head()
```

```python
train_features = train_df[['age','continent','position','prefer_foot','reputation','stat_overall','stat_potential','stat_skill_moves']]
train_features.head()
```

```python
def standard_scaling(df,cols) :
    for i in cols :
        mean_cols = df[i].mean()
        std_cols = df[i].std()
        df[i] = df[i].apply(lambda x : (x - mean_cols)/std_cols)
    
    return df        
```

```python
cols = ['age','continent','position','prefer_foot','reputation','stat_overall','stat_potential','stat_skill_moves']
standard_scaling(train_features,cols)
```

```python
train_features.head()
```

```python
target = np.log(train_df['value']).astype('int')
```

```python
print(len(target))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators = 100, random_state = 42)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(train_features,target, test_size = 0.2 , random_state = 42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
rf.fit(X_train,y_train)
```

```python
# 랜덤포레스트 변수별 중요도
for name, score in zip(train_features.columns, rf.feature_importances_) :
    print(name, score)
```

```python
y_pred = rf.predict(X_test)
```

```python
# 랜덤포레스트 정확도
from sklearn.metrics import accuracy_score 
accuracy_score(y_test,y_pred)
```

```python
XX = train_features[['age','continent','position','prefer_foot','reputation','stat_overall','stat_potential','stat_skill_moves']]
XX[:10].head()
```

```python
plt.figure(figsize = (12,8))
plt.plot(np.log(train_df['value'][:10]))
plt.plot(rf.predict(XX[:10]))
plt.show()
```

```python
param_grid = {'max_depth' : [None,2,3,4,5,6], 'criterion' : ['mse'], 'max_leaf_nodes' : [None,2,3]}
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV

dr = DecisionTreeRegressor()
grid = GridSearchCV(dr, param_grid,)
```

```python
target = train_df['value']
```

```python
X_train,X_test,y_train,y_test = train_test_split(train_features,target, test_size = 0.2)
```

```python
grid.fit(X_train,y_train)
```

```python
# 옵션중 가장 높은 score
print(grid.best_score_)
# 최상의 score일 때 옵션
print(grid.best_params_)
```

```python
grid.predict(X_test)
```
